```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, label="Boston", color = "navy", linewidth = 2, marker ="*")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("Pyre Fare by Month")
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, label = "Boston", color = "green", linewidth = 2, marker = "D")
ax.set_xlabel("Months")
ax.set_ylabel("Fare($)")
ax.grid()
ax.legend()
ax.set_title("PyBear Fare by Month")
```

```python
plt.barh(x_axis, y_axis, color = "turquoise", label = "Boston")
plt.title("PyBer")
plt.xlabel("Months")
plt.ylabel("Fare($)")
plt.legend()
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "coral", label = "Chicago")
ax.set_ylabel("Months")
ax.set_xlabel("Fare($)")
ax.set_title("Pyber Analysis")
```

```python
yaxis_larger = []
for data in y_axis:
    yaxis_larger.append(data * 3)
    
plt.scatter(y_axis, x_axis, c = "red", label = "Chicago", s = [i * 5 for i in y_axis])
plt.ylabel("Fare($)")
plt.xlabel("Months")
plt.title("PyBer Fare per month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c = "skyblue", linewidth = 2, edgecolor = "k", label = "Chicago")
ax.set_xlim(0,50)
ax.set_title ("plot verguita")
ax.set_xlabel("numebr")
ax.set_ylabel("meses bro")
plt.gca().invert_yaxis()
ax.legend()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.7, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
colors = ["coral", "turquoise", "lime", "lightblue", "red", "tomato", "skyblue", "blue", "brown", "darkred", "green", "yellow"]
plt.pie(y_axis, labels = x_axis, explode = explode_values, autopct = "%.1f%%", colors = colors)
```

```python
explode_values = (0, 0, 0.4, 0, 0, 0, 0.4, 0, 0, 0, 0, 0)
fig, ax = plt.subplots()
colors = ["darksalmon", "sandybrown","moccasin", "lightcoral", "coral", "indianred", "salmon", "darkseagreen","turquoise","thistle","yellowgreen","plum"]

ax.pie(y_axis, labels = x_axis, autopct = "%.1f%%", radius = 2.5, shadow = True, explode = explode_values, startangle = 90, colors=colors)
plt.gca().invert_xaxis()
```
